```python
df=df_merged[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
df.dropna(inplace=True)
```

```python
string=df.iloc[0].genres
```

```python
df.iloc[0].genres
```

```python
type(df.iloc[0].genres)
```

```python
ary_list=ast.literal_eval(string)
```

```python
ary_list
```

```python
ary_list[0]['name']
```

```python
def extract(obj):
    L=[]
    ary_list=ast.literal_eval(obj)
    for i in ary_list:
        L.append(i['name'])
    return L
```

```python
extract('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
df['genres']=df['genres'].apply(lambda x:extract(x))
```

```python
df['genres']
```

```python
df['keywords']=df['keywords'].apply(lambda x:extract(x))
```

```python
df['keywords']
```

```python
df.iloc[0].cast
```

```python
def extract_cast(obj):
    arr_list=ast.literal_eval(obj)
    counter=0
    L=[]
    for i in arr_list:
        if counter<3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
        
df['cast']=df['cast'].apply(lambda x:extract_cast(x))
```

```python
df['cast']
```

```python
df['crew']=df['crew'].apply(lambda x:[i['name'] for i in ast.literal_eval(x) if i['job'] == 'Director'])
```

```python
df['crew']
```

```python
df.head()
```

```python
type(df.iloc[0].overview)
```

```python
# ast.literal_eval(df.iloc[0].overview)
```

```python
df.iloc[0].overview
```

```python
df['overview']=df['overview'].apply(lambda x: x.split())
```

```python
df['overview']
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
# df['overview']=df['overview'].apply(collapse)
# there is no need because it is already split bu space
df['cast']=df['cast'].apply(collapse)
df['crew']=df['crew'].apply(collapse)
df['keywords']=df['keywords'].apply(collapse)
df['genres']=df['genres'].apply(collapse)
```

```python
df
```

```python
df['tags'] = df['overview'] + df['genres'] + df['keywords'] + df['cast'] + df['crew']
```

```python
df.iloc[0].tags
```

```python
df_new=df.drop(['crew','cast','overview','keywords','genres'],axis=1)
```

```python
df_new.head()
```

```python
df_new['tags']=df['tags'].apply(lambda x:" ".join(x))
```

```python
df_new.iloc[0]['tags']
```

```python
df_new['tags']=df_new['tags'].apply(lambda x:x.lower())
```

```python
df_new
```

```python
# import ntlk
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(df_new['tags']).toarray()
# vector will give us the class
```

```python
vector[0]
```

```python
vector.shape
# this tell us the sie of vector is the (no. of rowa,no. of feature)
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
similarity
```

```python
similarity.shape
```

```python
list(enumerate(similarity[0]))
```

```python
def recommend(movie):
    index=df[df['title']==movie].index[0]
    distance=similarity[index]
    rec_index=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])
    for i in rec_index[1:6]:
        print(df_new.iloc[i[0]].title)
        
```

```python
recommend("Batman")
```
