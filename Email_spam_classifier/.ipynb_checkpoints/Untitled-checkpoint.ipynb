```python
import string
from nltk.stem.porter import PorterStemmer
nltk.download('stopwords')
from nltk.corpus import stopwords
ps=PorterStemmer()
def transformed(text):
    y=[]
    text=text.lower()
    text=nltk.word_tokenize(text)
    for i in text:
        if i.isalnum():
            y.append(i)
    text=y[:]
    y.clear()
    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            y.append(ps.stem(i))
            
    return " ".join(y)
    
```

```python
transformed("hii hOW are YOU ? Vinki")
```

```python
df["transformed"]=df['msg'].apply(lambda x:transformed(x))
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
cv=CountVectorizer(max_features=3000)
#  also try with no max_features
tfid=TfidfVectorizer()
```

```python
vector=cv.fit_transform(df["transformed"]).toarray()
```

```python
vector
```

```python
# scale the vector
```

```python
y=df['output'].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score
x_train,x_test,y_train,y_test=train_test_split(vector,y,random_state=2,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
gs=GaussianNB()
mu=MultinomialNB()
br=BernoulliNB()
```

```python
gs.fit(x_train,y_train)
y_pred1=gs.predict(x_test)
print(accuracy_score(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
print(precision_score(y_test,y_pred1))
```

```python
mu.fit(x_train,y_train)
y_pred2 = mu.predict(x_test)
print(accuracy_score(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
print(precision_score(y_test,y_pred2))
```

```python
br.fit(x_train,y_train)
y_pred3 = br.predict(x_test)
print(accuracy_score(y_test,y_pred3))
print(confusion_matrix(y_test,y_pred3))
print(precision_score(y_test,y_pred3))
```
